```python
def matrix_cosine(merge_coll, top=200, cutoff=0.9):
    from scipy.spatial.distance import cosine

    stack = []
    res = dict()
    for i in merge_coll:
        res[i] = dict()
        for j in merge_coll:
            dist = 1 - cosine(merge_coll[i][:top].fillna(0), merge_coll[j][:top].fillna(0))  
            if dist >= cutoff:
                res[i][j] = dist
            else:
                res[i][j] = 0
    return res
```
